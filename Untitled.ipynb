```python
from google.cloud import bigquery

# Initialize a BigQuery client
client = bigquery.Client()

# Specify the dataset and table name
project_id = 'stone-resource-428206-q7'  # Replace with your actual project ID
dataset_id = 'your_dataset'     # Replace with your actual dataset name
table_id = 'cloudwatch_traffic_web_attack'  # Replace with your desired table name

# Define the URI of the file in Google Cloud Storage
uri = 'gs://your_bucket_name/CloudWatch_Traffic_Web_Attack.csv'  # Replace with your bucket and file name

# Configure the load job
job_config = bigquery.LoadJobConfig(
    source_format=bigquery.SourceFormat.CSV,
    skip_leading_rows=1,  # Adjust if your CSV has headers
    autodetect=True,  # Let BigQuery infer the schema
)

# Load the data
load_job = client.load_table_from_uri(
    uri, f'{project_id}.{dataset_id}.{table_id}', job_config=job_config
)

load_job.result()  # Wait for the job to complete
print("Data loaded successfully.")
```
